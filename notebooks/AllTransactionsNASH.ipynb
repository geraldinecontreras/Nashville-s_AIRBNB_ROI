```python
redfinsales = pd.read_csv('../data/redfin_2021-12-09-17-27-26.csv')
redfinsales.shape
```

```python
redfinsales.head()
```

```python
#drops columns that have only NAN values
redfinsales.dropna(how='all', axis=1, inplace=True)
```

```python
print(redfinsales.columns.tolist())
```

```python
#drop column
redfinsales = redfinsales.drop(['FAVORITE', 'INTERESTED', 'URL (SEE https://www.redfin.com/buy-a-home/comparative-market-analysis FOR INFO ON PRICING)'], axis=1)
```

```python
#drop another column
redfinsales = redfinsales.drop(['SALE TYPE'], axis=1)
```

```python
#rename columns
redfinsales = redfinsales.rename(columns={'STATE OR PROVINCE': 'State'})
redfinsales = redfinsales.rename(columns={'ZIP OR POSTAL CODE': 'Zipcode'})
redfinsales = redfinsales.rename(columns={'SQUARE FEET': 'Sqft2'})
redfinsales = redfinsales.rename(columns={'HOA/MONTH': 'HOA'})
redfinsales = redfinsales.rename(columns={'$/SQUARE FEET': 'Pricepersqft'})
```
